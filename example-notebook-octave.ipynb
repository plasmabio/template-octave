## Graphics

```octave
x=0:pi/100:2*pi;
y=sin(x);
y2 = sin(x-0.25);
y3=sin(x+0.5);
plot (x,y,'ok',x,y2,'--b',x,y3,'dm');
legend('sin(x)','sin(x-2.5)','sin(x-0.5)'); 
```

Other graphics taken from https://github.com/Calysto/octave_kernel

```octave
%plot --format png
tx = ty = linspace (-8, 8, 41)';
[xx, yy] = meshgrid (tx, ty);
r = sqrt (xx .^ 2 + yy .^ 2) + eps;
tz = sin (r) ./ r;
mesh (tx, ty, tz);
```

```octave
% Create an RGB image with no title or labels
close all hidden
image = zeros(300,400,3);
image(1:100,:,1) = 0.9;  % Red
image(101:200,:,2) = rand(100,400);  % Green
image(201:300,:,3) = 0.9;  % Blue
imshow(image)
```
